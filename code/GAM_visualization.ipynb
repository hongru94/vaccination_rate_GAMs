```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_all = pd.read_csv('/vaccination_rate_GAMs/results/model_1_all_waves.csv')
data_all = data_all.drop(columns = ['Unnamed: 0']).drop_duplicates().reset_index().drop(columns = ['index'])
```

```python
df_all = pd.read_csv('/vaccination_rate_GAMs/data/df_hosp.csv')
```

```python
var_name = ['Relative_completed_primary_series_rate_lag2', 'Relative_Past_12_week_infections_lag4', 'Relative_full_service_restaurant_lag2', 
            'Relative_gym_lag2',
            'Relative_office_of_physician_lag2', 'Relative_university_lag2', 'Relative_Government_Response_Index_lag2', 
            'Relative_Weekly_testing_lag2']
```

```python
var_label = ['Relative complete primary series rate', 'Relative previous infections (12 weeks)', 'Relative full-service restaurant visits', 'Relative gym visits',
            'Relative office of physician visits', 'Relative university visits', 'Relative government response index', 'Relative weekly testing rate']
```

```python
fig, axs = plt.subplots(3, 3, figsize = (16,12))
plt.tight_layout(pad=0.4, w_pad=3, h_pad=4.5)
fig.delaxes(axs[2,2])
plt.setp(axs, ylim=(-0.8, 0.6))

for i in [0, 1, 2]:
    for j in [0, 1, 2]:
        if i + j != 4:
            var_all = data_all[data_all['smooth'] == 's(' + var_name[i*3 + j] + ')'][var_name[i*3 + j]]
            est_all = data_all[(data_all['smooth'] == 's(' + var_name[i*3 + j] + ')')]['est']
            se_all = data_all[(data_all['smooth'] == 's(' + var_name[i*3 + j] + ')')]['se']
            
            axs[i,j].plot(var_all, est_all, color = 'forestgreen', label = 'All waves')
            axs[i,j].fill_between(var_all, est_all + 1.96*se_all, 
                      est_all - 1.96*se_all, color = 'forestgreen', alpha =.15)
            axs[i,j].legend()
            axs[i,j].set_ylabel('Estimated Partial Effects', fontsize = 15)
            axs[i,j].set_xlabel(var_label[i*3 + j], fontsize = 15)
            axs[i,j].set_title('s(' + var_label[i*3 + j] + ')', fontsize = 15)
            axs[i,j].plot((df_all[var_name[i*3 + j]], df_all[var_name[i*3 + j]]), 
                          (-1.58, -1.6), 'k-', color = 'forestgreen')
```

### Compare multiple waves

```python
data = pd.read_csv('/vaccination_rate_GAMs/results/model_1_compare_3_waves.csv').drop(columns = {'Unnamed: 0'})
```

```python
df_others = df_all[df_all['Dominant_variant'] == 'Pre-Delta'].reset_index().drop(columns = ['index'])
df_delta = df_all[df_all['Dominant_variant'] == 'Delta'].reset_index().drop(columns = ['index'])
df_omicron = df_all[df_all['Dominant_variant'] == 'Omicron'].reset_index().drop(columns = ['index'])
```

```python
fig, axs = plt.subplots(3, 3, figsize = (16,12))
plt.tight_layout(pad=0.4, w_pad=3, h_pad=4.5)
fig.delaxes(axs[2,2])
plt.setp(axs, ylim=(-1, 0.5))

for i in [0, 1, 2]:
    for j in [0, 1, 2]:
        if i + j != 4:
            var_others = data[(data['model'] == 'model_1_pre_delta')&(data['smooth'] == 's(' + var_name[i*3 + j] + ')')][var_name[i*3 + j]]
            var_delta = data[(data['model'] == 'model_1_delta')&(data['smooth'] == 's(' + var_name[i*3 + j] + ')')][var_name[i*3 + j]]
            var_omicron = data[(data['model'] == 'model_1_omicron')&(data['smooth'] == 's(' + var_name[i*3 + j] + ')')][var_name[i*3 + j]]

            est_others = data[(data['model'] == 'model_1_pre_delta')&(data['smooth'] == 's(' + var_name[i*3 + j] + ')')]['est']
            est_delta = data[(data['model'] == 'model_1_delta')&(data['smooth'] == 's(' + var_name[i*3 + j] + ')')]['est']
            est_omicron = data[(data['model'] == 'model_1_omicron')&(data['smooth'] == 's(' + var_name[i*3 + j] + ')')]['est']

            se_others = data[(data['model'] == 'model_1_pre_delta')&(data['smooth'] == 's(' + var_name[i*3 + j] + ')')]['se']
            se_delta = data[(data['model'] == 'model_1_delta')&(data['smooth'] == 's(' + var_name[i*3 + j] + ')')]['se']
            se_omicron = data[(data['model'] == 'model_1_omicron')&(data['smooth'] == 's(' + var_name[i*3 + j] + ')')]['se']

            axs[i,j].plot(var_others, est_others, color = 'deepskyblue', label = 'Pre-Delta Wave')
            axs[i,j].plot(var_delta, est_delta, color = 'orange', label = 'Delta Wave')
            axs[i,j].plot(var_omicron, est_omicron, color = 'crimson', label = 'Omicron Wave')

            axs[i,j].fill_between(var_others, est_others + 1.96*se_others, 
                      est_others - 1.96*se_others, color = 'deepskyblue', alpha =.15)
            axs[i,j].fill_between(var_delta, est_delta + 1.96*se_delta, 
                      est_delta - 1.96*se_delta, color = 'orange', alpha =.15)
            axs[i,j].fill_between(var_omicron, est_omicron + 1.96*se_omicron, 
                      est_omicron - 1.96*se_omicron, color = 'crimson', alpha =.15)
            axs[i,j].legend()
            axs[i,j].set_ylabel('Estimated Partial Effects', fontsize = 15)
            axs[i,j].set_xlabel(var_label[i*3 + j], fontsize = 15)
            axs[i,j].set_title('s(' + var_label[i*3 + j] + ')', fontsize = 15)

            axs[i,j].plot((df_others[var_name[i*3 + j]], df_others[var_name[i*3 + j]]), (-0.98, -1), 'k-', color = 'deepskyblue', alpha = 0.5)
            axs[i,j].plot((df_delta[var_name[i*3 + j]], df_delta[var_name[i*3 + j]]), (-0.98, -1), 'k-', color = 'orange', alpha = 0.5)
            axs[i,j].plot((df_omicron[var_name[i*3 + j]], df_omicron[var_name[i*3 + j]]), (-0.98, -1), 'k-', color = 'crimson', alpha = 0.5)







```
